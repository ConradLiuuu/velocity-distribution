```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data(name):
    path = './fixed classification/' + name + '_plt.csv'
    data = pd.read_csv(path, header=None)
    data = np.array(data)
    
    vx = data[:,0]
    vy = data[:,1]
    vz = data[:,2]
    theta = data[:,3]
    v = data[:,4]
    
    return vx, vy, vz, theta, v
```

```python
top5_vx, top5_vy, top5_vz, top5_theta, top5_v = read_data('top5')
top6_vx, top6_vy, top6_vz, top6_theta, top6_v = read_data('top6')

left5_vx, left5_vy, left5_vz, left5_theta, left5_v = read_data('left5')
left6_vx, left6_vy, left6_vz, left6_theta, left6_v = read_data('left6')

right5_vx, right5_vy, right5_vz, right5_theta, right5_v = read_data('right5')
right6_vx, right6_vy, right6_vz, right6_theta, right6_v = read_data('right6')

back5_vx, back5_vy, back5_vz, back5_theta, back5_v = read_data('back5')
back6_vx, back6_vy, back6_vz, back6_theta, back6_v = read_data('back6')
```
