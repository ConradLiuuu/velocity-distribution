```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data(name):
    path = './fixed classification/' + name + '.csv'
    data = pd.read_csv(path, header=None)
    data = np.array(data)
    
    vx = data[:,0]
    vy = data[:,1]
    vz = data[:,2]
    theta = data[:,3]
    v = data[:,4]
    
    return vx, vy, vz, theta, v
```

```python
vel5_vx, vel5_vy, vel5_vz, vel5_theta, vel5_v = read_data('vel5')
vel6_vx, vel6_vy, vel6_vz, vel6_theta, vel6_v = read_data('vel6')
```

```python
vel = [vel5_v, vel6_v]
```

```python
plt.figure(figsize=(20,10))

boxplot_dict = plt.boxplot(vel)
for i in boxplot_dict['medians']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
    #print(y)
for i in boxplot_dict['boxes']:
    x, y = i.get_xydata()[1]
    plt.text(x, y, '%.3f' % y,horizontalalignment='left',verticalalignment='top')
    x2, y2 = i.get_xydata()[2]
    plt.text(x2, y2, '%.3f' % y2,horizontalalignment='left',verticalalignment='bottom')
    #print(y)
for i in boxplot_dict['caps']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
    #print(y)
plt.xticks([1,2],['vel5', 'vel6'])
#plt.ylabel('vx')
plt.title('V')
plt.tight_layout()
plt.savefig('./fixed plot_20200708/fixed all data vel.png')
plt.savefig('./fixed plot_20200708/fixed all data vel.svg')
```
