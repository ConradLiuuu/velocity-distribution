```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tt = pd.read_csv('./tt.csv', header=None)
top5 = np.array(tt)
top5.shape
```

```python
tt2 = pd.read_csv('./tt2.csv', header=None)
top6 = np.array(tt2)
top6.shape
```

```python
ttttt = [top5, top6]
```

```python
plt.figure(figsize=(10,10))

boxplot_dict = plt.boxplot(ttttt)
for i in boxplot_dict['medians']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
    #print(y)
for i in boxplot_dict['boxes']:
    x, y = i.get_xydata()[1]
    plt.text(x, y, '%.3f' % y,horizontalalignment='left',verticalalignment='top')
    x2, y2 = i.get_xydata()[2]
    plt.text(x2, y2, '%.3f' % y2,horizontalalignment='left',verticalalignment='bottom')
    #print(y)
for i in boxplot_dict['caps']:
    x, y = i.get_xydata()[0]
    plt.text(x, y, '%.3f' % y,horizontalalignment='right',verticalalignment='center')
    #print(y)
plt.xticks([1,2],['top5', 'top6'])
#plt.ylabel('vx')
plt.tight_layout()
plt.title('az/vy')
```

```python
def read_data(name):
    path = './fixed classification/' + name + '_plt.csv'
    data = pd.read_csv(path, header=None)
    data = np.array(data)
    
    vx = data[:,0]
    vy = data[:,1]
    vz = data[:,2]
    theta = data[:,3]
    v = data[:,4]
    
    return vx, vy, vz, theta, v
```

```python
top5_vx, top5_vy, top5_vz, top5_theta, top5_v = read_data('top5')
top6_vx, top6_vy, top6_vz, top6_theta, top6_v = read_data('top6')
```

```python
plt.scatter(top5_theta, top5_v, color='b')
plt.scatter(top6_theta, top6_v, color='r')
```
